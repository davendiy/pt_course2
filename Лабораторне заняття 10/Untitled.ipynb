```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from scipy.stats import bernoulli

p_list = [0.05, 0.25, 0.5, 0.75, 0.95]
n_list = np.arange(1, 1001)

for p in p_list:
    tmp_rv = bernoulli(p)
    rvs = tmp_rv.rvs(1000)
    rvs = rvs.cumsum()
    plt.plot(n_list, rvs / n_list, label=f'p = {p}')
    plt.hlines(p, 0, 1000)
#     plt.legend(loc='best')
    plt.ylim(0, 1)
#     plt.show()
plt.legend(loc='best')
```

```python
from scipy.stats import expon

p_list = [1, 0.5, 2, 5]
n_list = np.arange(1, 1001)



for p in p_list:
    tmp_rv = expon(0, 1/p)
    rvs = tmp_rv.rvs(1000)
    rvs = rvs.cumsum()             # np.arrray.cumsum
    plt.plot(n_list, rvs / n_list, label=f'p = {p}')
    plt.hlines(1/p, 0, 1000)
#     plt.legend(loc='best')
#     plt.show()
plt.legend(loc='best')
```

```python
from scipy.stats import norm

rv = norm()

rvs = np.abs(rv.rvs(100000))

res = np.maximum.accumulate(rvs) / np.sqrt(np.arange(1, 100001))
plt.plot(res)
plt.hlines(0, 1, 100000)
```

```python
from scipy.stats import uniform

def f(x): 
    return np.sin(x)


n = 10000
x = f(uniform(0, 1).rvs(n))
plt.plot(x.cumsum() / np.arange(1, n+1))
plt.ylim(0, 0.6)
# plt.hlines(1/3, 0, 10000)
```

```python
from scipy.stats import expon

def f(x):
    return 1/x

n = 10000000
x = expon.rvs(size=n, loc=1) 
x = f(x) / expon.pdf(x, loc=1)
plt.plot(x.cumsum() / np.arange(1, n+1))
```
